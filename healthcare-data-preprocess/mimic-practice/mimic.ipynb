```python
import pandas as pd
base_path='/Users/minha/ai/ai-practice/healthcare-data-preprocess/mimic-iii-clinical-database-demo-1.4/'

patients =  pd.read_csv(f"{base_path}PATIENTS.csv", index_col=0)
print( f"(레코드 수, 속성 수): {patients.shape}")
print(patients.info())
print(patients.head())
```

```python
patients.head()
```

```python
# 모든 patient 테이블 설명
patients.describe(include='all')
```

```python
# patient 테이블의 dob 컬럼을 datetime으로 바꿔서 추출
dob = patients.dob.astype('datetime64[ms]')
dob.dtype
```

```python
patients.dob = dob
date_from = pd.to_datetime('2000-01-01')
date_to = pd.to_datetime('2200-12-31')
patients[(date_from < patients.dob ) & (patients.dob < date_to)]
```

```python
patients.info()
```

```python
# gender 컬럼 갯수
patients.gender.value_counts()
```

```python
# ununique한 컬럼들 갯수
patients.nunique()
```

```python
# subject_id 부터 dob 까지의 값 추출
patients.loc[:, 'subject_id':'dob']
```

```python
patients.to_sql(name='PATIENTS', con=con, if_exists='replace', index=False )
a = pd.read_sql( ' select * from PATIENTS', con=con)
print(a.head())
```
